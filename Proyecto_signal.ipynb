```python
import os
```

```python

# Directorio raíz que contiene las carpetas con imágenes
directorio_raiz = "/content/archiveseñas"

# Extensión de las imágenes (puedes cambiarlo según las extensiones de tus imágenes)
extension = ".jpg"

# Diccionario para almacenar la cantidad de imágenes por carpeta
imagenes_por_carpeta = {"train": 0, "test": 0}

# Función para contar imágenes de forma recursiva y separar por carpeta
def contar_imagenes(directorio):
    cantidad_imagenes = 0
    for root, _, files in os.walk(directorio):
        cantidad_en_carpeta = len([archivo for archivo in files if archivo.endswith(extension)])
        cantidad_imagenes += cantidad_en_carpeta

        # Verificar si la carpeta actual está dentro de "train" o "test"
        if "train" in root:
            imagenes_por_carpeta["train"] += cantidad_en_carpeta
        elif "test" in root:
            imagenes_por_carpeta["test"] += cantidad_en_carpeta

    return cantidad_imagenes

# Contar todas las imágenes en el directorio raíz y obtener la cantidad por carpeta
cantidad_total_imagenes = contar_imagenes(directorio_raiz)

# Imprimir la cantidad de imágenes por carpeta
for carpeta, cantidad in imagenes_por_carpeta.items():
    print(f"Total de imágenes en la carpeta '{carpeta}': {cantidad}")

# Imprimir el total de imágenes en todas las carpetas
print(f"Total de imágenes en la carpeta raíz y sus subcarpetas: {cantidad_total_imagenes}")
```

```python
import matplotlib.pyplot as plt
from PIL import Image

directorio_raiz = "/content/archiveseñas"
extension = ".jpg"
imagenes_por_carpeta = {"train": None, "test": None}

def contar_imagenes(directorio):
    cantidad_imagenes = 0
    for root, _, files in os.walk(directorio):
        cantidad_en_carpeta = len([archivo for archivo in files if archivo.endswith(extension)])
        cantidad_imagenes += cantidad_en_carpeta

        if "train" in root and imagenes_por_carpeta["train"] is None:
            imagenes_por_carpeta["train"] = os.path.join(root, files[0]) if cantidad_en_carpeta > 0 else None
        elif "test" in root and imagenes_por_carpeta["test"] is None:
            imagenes_por_carpeta["test"] = os.path.join(root, files[0]) if cantidad_en_carpeta > 0 else None

    return cantidad_imagenes

cantidad_total_imagenes = contar_imagenes(directorio_raiz)

for carpeta, imagen in imagenes_por_carpeta.items():
    print(f"Imagen representativa de la carpeta '{carpeta}':")
    if imagen is not None:
        print(imagen)
        img = Image.open(imagen)
        plt.imshow(img)
        plt.axis('off')  # Ocultar ejes
        plt.show()
    else:
        print("No se encontraron imágenes en esta carpeta.")

print(f"Total de imágenes en la carpeta raíz y sus subcarpetas: {cantidad_total_imagenes}")
```
